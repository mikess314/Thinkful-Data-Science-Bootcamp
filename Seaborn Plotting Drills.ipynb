-Choose two continuous variables, and plot them three different ways.

```python
sns.set(style="darkgrid")
g = sns.factorplot(x="manner_of_death", y="id", data=killings, size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Police Fatalities")
g.set_xlabels("")
plt.title('Police Killings by Manner of Death')
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.set(style="whitegrid", palette="pastel", color_codes=True)
sns.violinplot(x="race", y="age", hue="gender", data=killings, split=True)
sns.despine(left=True)
```

```python
sns.set(style="whitegrid")
g = sns.factorplot(x="manner_of_death", y="id", data=killings, size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Police Fatalities")
g.set_xlabels("Manner of Death")
plt.title('Police Killings by Manner of Death')
plt.show()
```

```python
plt.figure(figsize=(6,12))
sns.set(style="ticks")
sns.boxplot(x="manner_of_death", y="age", data=killings, palette="PRGn")
sns.despine(offset=10, trim=True)
```
